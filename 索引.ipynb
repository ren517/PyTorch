```python
import torch
```

```python
torch.manual_seed(111)
t1 = torch.randint(low=0, high=10,size=(5,5))
print(t1)
```

### 返回（0,1）（1,2）

```python
print(t1[[0, 1],[1, 2]])
```

### 返回（1, 2）（2， 4）

```python
print(t1[[1, 3],[2, 4]])
```

### 取0,1行的第1,2列四个元素

```python
print(t1[:2,1:3])
```

### 所有奇数行， 偶数行

```python
print(t1[::2, :])
print(t1[1::2, :])
```

### 所有奇数行， 偶数列

```python
print(t1[::2, 1::2])
```

### bool 类型

```python
print(t1[torch.tensor([True, False, True, False, True]), :])
```

```python
# 第三列所有大于五的数字
print(t1[:,t1[: ,2] > 5])
```
